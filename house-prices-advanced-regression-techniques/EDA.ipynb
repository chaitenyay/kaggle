## Handle Missing Data

### Categorical Features

```python
combineData.Alley = combineData.Alley.fillna('None')
combineData.PoolQC = combineData.PoolQC.fillna('None')
combineData.FireplaceQu = combineData.FireplaceQu.fillna('None')
combineData.GarageCond = combineData.GarageCond.fillna('None')
combineData.GarageQual = combineData.GarageQual.fillna('None')
combineData.GarageFinish = combineData.GarageFinish.fillna('None')
combineData.GarageType = combineData.GarageType.fillna('None')
combineData.BsmtCond = combineData.BsmtCond.fillna('None')
combineData.BsmtExposure = combineData.BsmtExposure.fillna('None')
combineData.BsmtQual = combineData.BsmtQual.fillna('None')
combineData.BsmtFinType2 = combineData.BsmtFinType2.fillna('None')
combineData.BsmtFinType1 = combineData.BsmtFinType1.fillna('None')
combineData.MasVnrType = combineData.MasVnrType.fillna('None')
combineData.MiscFeature = combineData.MiscFeature.fillna('None')
combineData.Fence = combineData.Fence.fillna('None')


combineData.MSZoning = combineData.MSZoning.fillna(combineData.MSZoning.mode()[0])
combineData.Utilities = combineData.Utilities.fillna(combineData.Utilities.mode()[0])
combineData.Functional = combineData.Functional.fillna(combineData.Functional.mode()[0])
combineData.KitchenQual = combineData.KitchenQual.fillna(combineData.KitchenQual.mode()[0])
combineData.SaleType = combineData.SaleType.fillna(combineData.SaleType.mode()[0])
combineData.Exterior2nd = combineData.Exterior2nd.fillna(combineData.Exterior2nd.mode()[0])
combineData.Exterior1st = combineData.Exterior1st.fillna(combineData.Exterior1st.mode()[0])
combineData.Electrical = combineData.Electrical.fillna(combineData.Electrical.mode()[0])
```

### Numeric Features

```python
combineData.LotFrontage = combineData.groupby('Neighborhood').LotFrontage.transform(lambda x: x.fillna(x.median()))
```

```python
combineData.GarageArea = combineData.GarageArea.fillna(0)
combineData.GarageCars = combineData.GarageCars.fillna(0)
combineData.GarageYrBlt = combineData.GarageYrBlt.fillna(0)

combineData.MasVnrArea = combineData.MasVnrArea.fillna(0)

combineData.BsmtFinSF1 = combineData.BsmtFinSF1.fillna(0)
combineData.BsmtFinSF2 = combineData.BsmtFinSF2.fillna(0)
combineData.BsmtUnfSF = combineData.BsmtUnfSF.fillna(0)
combineData.TotalBsmtSF = combineData.TotalBsmtSF.fillna(0)
combineData.BsmtHalfBath = combineData.BsmtHalfBath.fillna(0)
combineData.BsmtFullBath = combineData.BsmtFullBath.fillna(0)
```
